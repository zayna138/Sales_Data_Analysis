```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Product': ['Laptop', 'Phone', 'Tablet', 'Headphones', 'Monitor'],
    'Sales': [50000, 40000, 25000, 15000, 30000],
    'Profit': [10000, 8000, 5000, 3000, 6000]
}

df = pd.DataFrame(data)

print("Sales Data")
print(df)

top_product = df[df['Sales'] == df['Sales'].max()]
print("\nHighest Selling Product:")
print(top_product[['Product', 'Sales']])

plt.figure(figsize=(8,5))
plt.bar(df['Product'], df['Sales'])
plt.xlabel("Products")
plt.ylabel("Sales")
plt.title("Product Sales Analysis")
plt.show()
```
